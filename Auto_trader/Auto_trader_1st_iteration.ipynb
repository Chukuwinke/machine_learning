```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#READING FILE FROM CSV
auto_data = pd.read_csv('./Resources/adverts.csv')
```

```python
# MAKING A COPY OF THE FILE AND ADDING AND TAKIN ONLY A SAMPLE OF IT
auto_dat_copy = auto_data.copy().sample(n=5000, ignore_index=True)
auto_dat_copy.shape
```

```python
auto_dat_copy.isnull().sum()
```

```python
auto_dat_copy.head(5)
```

```python
f_col = auto_dat_copy.columns
for x in f_col:
    dc = auto_dat_copy[x]
    if auto_dat_copy[x].dtypes =='O':
        dc_m = dc.fillna(auto_dat_copy[x].mode().iloc[0])
    else:
        dc_m = dc.fillna(value = auto_dat_copy[x].mean())
    auto_dat_copy.update({x: dc_m})
```

```python
auto_dat_copy.isnull().sum()
```

```python
auto_dat_copy.head(60)
```

```python
data_numeric = auto_dat_copy[['mileage', 'vehicle_condition', 'year_of_registration', 'price']]
data_numeric.head(5)
```

```python
for column in data_numeric.columns:
    print(data_numeric[column].value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()

feature_transformed = ohe.fit_transform(data_numeric[["vehicle_condition"]]).toarray()
feature_labels = ohe.categories_[0]
```

```python
feature_labels = np.array(feature_labels).ravel()
```

```python
features_encoded = pd.DataFrame(feature_transformed, columns=feature_labels)
features_encoded.head(10)
```

```python
data_numeric = pd.concat([data_numeric, features_encoded], axis=1)
data_numeric.tail(10)
```
