```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
auto_data = pd.read_csv('../adverts.csv')
```

```python
auto_dat_copy = auto_data.copy().sample(n=5000)
auto_dat_copy.shape
```

```python
auto_dat_copy.isnull().sum()
```

```python
#auto_dat_copy.dtypes
```

```python
auto_dat_copy.head(50)
```

```python
f_col = auto_dat_copy.columns
for x in f_col:
    dc = auto_dat_copy[x]
    if auto_dat_copy[x].dtypes =='O':
        dc_m = dc.fillna(auto_dat_copy[x].mode().iloc[0])
    else:
        dc_m = dc.fillna(value = auto_dat_copy[x].mean())
    auto_dat_copy.update({x: dc_m})
```
